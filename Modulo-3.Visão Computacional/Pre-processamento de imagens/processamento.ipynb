```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread('imagens/img-passaro/pexels-roshan-kamath-1661179.jpg')
imagem = cv2.cvtColor(imagem,cv2.COLOR_BGR2RGB)
plt.imshow(imagem)
plt.show()
```

```python
imagem.shape
```

```python
print(f'\
valor maximo :{imagem.max()}           \n\
valor minimo :{imagem.min()}            \n\
media        :{imagem.mean()}            \n\
')
```

```python
b, g, r = cv2.split(imagem)

plt.figure(figsize=(10,12))

plt.subplot(3,3,2)
plt.imshow(imagem)

plt.subplot(3,3,4)
plt.imshow(b)

plt.subplot(3,3,5)
plt.imshow(g)

plt.subplot(3,3,6)
plt.imshow(r)

plt.subplot(3,3,7)
plt.imshow(b,cmap='gray')

plt.subplot(3,3,8)
plt.imshow(g,cmap='gray')

plt.subplot(3,3,9)
plt.imshow(r,cmap='gray')
plt.show()
```

```python
imagemR = cv2.resize(src=imagem,dsize=(50,300))
plt.imshow(imagemR)
plt.show()
```
