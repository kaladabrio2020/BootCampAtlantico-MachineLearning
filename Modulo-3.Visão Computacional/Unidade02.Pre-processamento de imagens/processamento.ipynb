```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread('imagens/img-passaro/pexels-roshan-kamath-1661179.jpg')
imagem = cv2.cvtColor(imagem,cv2.COLOR_BGR2RGB)
plt.imshow(imagem)
plt.show()
```

```python
imagem.shape
```

```python
print(f'\
valor maximo :{imagem.max()}           \n\
valor minimo :{imagem.min()}            \n\
media        :{imagem.mean()}            \n\
')
```

```python
b, g, r = cv2.split(imagem)

plt.figure(figsize=(10,12))

plt.subplot(3,3,2)
plt.imshow(imagem)

plt.subplot(3,3,4)
plt.imshow(b)

plt.subplot(3,3,5)
plt.imshow(g)

plt.subplot(3,3,6)
plt.imshow(r)

plt.subplot(3,3,7)
plt.imshow(b,cmap='gray')

plt.subplot(3,3,8)
plt.imshow(g,cmap='gray')

plt.subplot(3,3,9)
plt.imshow(r,cmap='gray')
plt.show()
```

```python
imagemR = cv2.resize(src=imagem,dsize=(50,300))
plt.imshow(imagemR)
plt.show()
```

```python
grayimage = cv2.cvtColor(imagem,cv2.COLOR_BGR2GRAY)
plt.imshow(grayimage,cmap='gray')
```

```python
imagemZero = cv2.imread('imagens/img-passaro/pexels-roshan-kamath-1661179.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(imagemZero,cmap='gray')
plt.show()
```

```python
gaussinb = cv2.GaussianBlur(imagemZero,(11,11),sigmaX=6)
plt.imshow(gaussinb,cmap='gray')
plt.show()
```

```python
plt.imshow(
    cv2.cvtColor(cv2.blur(imagem,(11,11)),cv2.COLOR_BGR2GRAY),cmap='gray'
)
plt.show()
```

```python
scale = cv2.convertScaleAbs(imagemZero,alpha=1.5,beta=100)
plt.imshow(scale,cmap='gray')
plt.show()
```

```python

hist = cv2.calcHist(imagemZero,[0],None, [256], [0, 256])

histimage = cv2.equalizeHist(imagemZero)

histhistimage = cv2.calcHist(histimage,[0],None, [256], [0, 256])


plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.imshow(imagemZero,cmap='gray')

plt.subplot(2,2,3)
plt.plot(hist)


plt.subplot(2,2,2)
plt.imshow(histimage,cmap='gray')

plt.subplot(2,2,4)
plt.plot(histhistimage)


plt.show()
```
