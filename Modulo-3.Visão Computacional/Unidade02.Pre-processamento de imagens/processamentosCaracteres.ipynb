```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imageP = cv2.imread('imagens/img-caracteres/pexels-magda-ehlers-1339865.jpg')
```

#### colorir imagem

```python
image = cv2.cvtColor(imageP,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

#### Detectar bordas

```python
imageCinza = cv2.cvtColor(imageP , cv2.COLOR_BGR2GRAY)
plt.imshow(imageCinza,cmap='gray')
plt.show()
```

```python
imageBorda = cv2.Canny(imageCinza,250,150)
plt.imshow(imageBorda,'gray')
plt.show()
```

##### 1. forma

```python
clache = cv2.convertScaleAbs(imageBorda,alpha=100,beta=1)
plt.imshow(clache,cmap='gray')
plt.show()
```

##### 2. forma 

```python
clache      = cv2.createCLAHE(2.0,(8,8))
imageClache = clache.apply(imageCinza) 


sobelx = cv2.Sobel(imageClache,cv2.CV_64F,1,0,ksize=9)
sobely = cv2.Sobel(imageClache,cv2.CV_64F,0,1,ksize=9)



imageGray = np.sqrt(sobelx**2 + sobely**2)


plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(imageClache,'gray')

plt.subplot(1,2,2)
plt.imshow(imageGray,'gray')

plt.show()
```

```python
gaussian = cv2.GaussianBlur(imageCinza,(101,101),sigmaX=12)
gaussian = cv2.medianBlur(gaussian,21)
sobelx = cv2.Sobel(gaussian,cv2.CV_64F,1,0,ksize=11)
sobely = cv2.Sobel(gaussian,cv2.CV_64F,0,1,ksize=11)



imageGray = np.sqrt(sobelx**2 + sobely**2)


plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.imshow(gaussian,'gray')

plt.subplot(2,2,2)
plt.imshow(imageGray,'gray')

plt.subplot(2,2,3)
imageBorda = cv2.Canny(gaussian,7,7)
plt.imshow(imageBorda,'gray')
plt.show()
```
