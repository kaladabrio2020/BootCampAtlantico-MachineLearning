1. Utilizando o matplotlib, crie e mostre um gráfico de linha simples e adicione títulos aos eixos X e Y do gráfico.

```````python
import matplotlib.pyplot as plt
from random import randint  


fig , ax = plt.subplots(figsize=(10,8))
ax.plot(
    [ randint(50,80)*i    for i in range(1,10)],
    [(randint(10,20)*i)-1 for i in range(1,10)],
    linestyle='--',
    color    ='red'
)
ax.grid(True)
ax.set_ylabel('valores y') # titulo eixo y
ax.set_xlabel('valores x') # titulo eixo x
ax.set_title('Plotagem hello')
plt.show()
```````

5. Utilizando pandas, como realizar a leitura de um arquivo CSV em um DataFrame e exibir as primeiras linhas?

```````python
# No csv tem esses dados
#"nome","idade"
#"Martins pereia",26
#"Amanda caroline",46
#"Catia nunes",120
#"tortuga taruga",32

import pandas as pd

dataset = pd.read_csv(r'dataset/exemplo.csv',sep=',')

print(dataset.shape)

print()
print(dataset.iloc[[0],:].to_string(index=False)) # exibindo linha 1

print()
print(dataset.iloc[0:2,:]) # exibindo linhas 1 e 2
```````
