1. Utilizando o matplotlib, crie e mostre um gráfico de linha simples e adicione títulos aos eixos X e Y do gráfico.

```````python
import matplotlib.pyplot as plt
from random import randint  


fig , ax = plt.subplots(figsize=(10,8))
ax.plot(
    [ randint(50,80)*i    for i in range(1,10)],
    [(randint(10,20)*i)-1 for i in range(1,10)],
    linestyle='--',
    color    ='red'
)
ax.grid(True)
ax.set_ylabel('valores y') # titulo eixo y
ax.set_xlabel('valores x') # titulo eixo x
ax.set_title('Plotagem hello')
plt.show()
```````

7. Utilizando pandas, como selecionar uma coluna específica e filtrar linhas em um “DataFrame” com base em uma condição?

```````python
dicionario = dict(
    nome=[    
        "Martins pereia",
        "Amanda caroline", 
        "Catia nunes",
        "tortuga taruga",
        "Carmo",
        "Katia numees",
        "tonali",
        "Carmen machine",
        "Cad"
        ],
    idade = [
        19,
        31,
        40,
        60,
        70,
        43,
        65,
        54,
        90
        ]
        )
import pandas as pd

dataset = pd.DataFrame(dicionario)

#primirio caractere do nome igual a 'C' com idade maior que 50
print(
    dataset['nome'].loc[
        (dataset['nome'].str.find('C',0,1) != -1) &
        (dataset['idade']>50)
    ]
)
```````

8. Utilizando pandas, como lidar com valores ausentes (NaN) em um DataFrame?

```````python
# removendo 
import numpy as np
import pandas as pd
dicionario = dict(
    nome=[    
        "Martins pereia","Amanda caroline",  "Katia numees", "tonali", "Carmen machine", "Cad"
    ],
    idade = [ 19, 31, 43, 90 ,np.NaN,np.NAN]    
)

dataset  = pd.DataFrame(dicionario)
print(dataset,'\n\n')

dataset.drop(
    dataset.loc[dataset['idade'].isnull()].index,inplace=True
)
print(dataset)
```````
