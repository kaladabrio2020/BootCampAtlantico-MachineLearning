```python
import numpy as np
import random
import matplotlib.pyplot as plt 
```

```python
x = np.array([random.randint(-5,5)*i for i in range(10)])
y = np.array([random.randint(0,10)*i for i in range(10)])
c = np.array([random.randint(10,100)*i for i in range(10)])

fig , ax = plt.subplots()

ax.scatter(
    x = x,
    y = y,
    s = c
)
fig.legend(
    'oi',
    shadow = True  ,
)
ax.set_ylabel('oidada',fontdict={
    'backgroundcolor':'red'
})
ax.set_xlabel('oidada')

ax.yaxis.set_label_position('right')
plt.grid(True)
fig.show()
```
